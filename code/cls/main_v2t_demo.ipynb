```python
import tkinter as tk
from tkinter import scrolledtext
import markdown
from tkhtmlview import HTMLLabel
from threading import Thread
import cls_GeneralTools
import cls_VoiceManager
import json
import pyaudio
import time

class StreamDisplayApp:
    def __init__(self, root, agentManager, test_mode = False):
        self.test_mode = test_mode
        self.root = root
        self.root.title("AI TRPG Manager by Gustaf Guo")
        self.root.geometry("1000x600")
        self.agentManager = agentManager
        self.voiceManager = cls_VoiceManager.VoiceManager()
        self.streaming = False
        self.full_response = ""
        self.history = []
        self.setup_ui()
        self.window_init()
        self.root.protocol("WM_DELETE_WINDOW", self.on_window_close)

        #Voice 2 text
        self.voice_recognition_active = False  # æ–°å¢è¯­éŸ³è¯†åˆ«çŠ¶æ€æ ‡å¿—
        self.vosk_model = None  # Voskæ¨¡å‹
        self.vosk_recognizer = None  # Voskè¯†åˆ«å™¨
        self.setup_ui()
        self.window_init()
        self.init_vosk()  # åˆå§‹åŒ–Vosk
        #pip install vosk -i https://pypi.tuna.tsinghua.edu.cn/simple --trusted-host pypi.tuna.tsinghua.edu.cn

    def init_vosk(self):
        """åˆå§‹åŒ–Voskè¯­éŸ³è¯†åˆ«"""
        try:
            from vosk import Model, KaldiRecognizer

            
            # æ£€æŸ¥æ¨¡å‹è·¯å¾„ï¼ˆéœ€æå‰ä¸‹è½½ä¸­æ–‡æ¨¡å‹ï¼‰
            model_path = "./vosk-model-small-cn-0.22"
            if not os.path.exists(model_path):
                print(f"è­¦å‘Šï¼šæœªæ‰¾åˆ°Voskä¸­æ–‡æ¨¡å‹ï¼Œè¯­éŸ³è¯†åˆ«ä¸å¯ç”¨ã€‚è¯·ä¸‹è½½å¹¶è§£å‹åˆ° {model_path}")
                return
                
            self.vosk_model = Model(model_path)
            self.vosk_recognizer = KaldiRecognizer(self.vosk_model, 16000)
            
            # åˆå§‹åŒ–éº¦å…‹é£
            self.pyaudio = pyaudio.PyAudio()
            self.audio_stream = self.pyaudio.open(
                format=pyaudio.paInt16,
                channels=1,
                rate=16000,
                input=True,
                frames_per_buffer=8192,
                stream_callback=self.audio_callback  # éŸ³é¢‘å›è°ƒå‡½æ•°
            )
            self.audio_stream.stop_stream()  # åˆå§‹æ—¶æš‚åœ
            
        except Exception as e:
            print(f"Voskåˆå§‹åŒ–å¤±è´¥: {e}")
            self.vosk_model = None      


    def audio_callback(self, in_data, frame_count, time_info, status):
        """éŸ³é¢‘æ•°æ®å›è°ƒå‡½æ•°"""
        if self.voice_recognition_active and self.vosk_recognizer:
            if self.vosk_recognizer.AcceptWaveform(in_data):
                result = json.loads(self.vosk_recognizer.Result())
                text = result.get("text", "")
                if text:
                    self.root.after(0, self.append_to_input, text + " ")
        return (None, pyaudio.paContinue)

    def setup_ui(self):
        # é…ç½®æ ¹çª—å£ç½‘æ ¼å¸ƒå±€
        self.root.grid_columnconfigure(0, weight=1)
        self.root.grid_columnconfigure(1, weight=1)
        self.root.grid_rowconfigure(0, weight=1)
        self.root.grid_rowconfigure(1, weight=0)  # çŠ¶æ€æ è¡Œ
        self.root.grid_rowconfigure(2, weight=0)  # å¤é€‰æ¡†è¡Œ

        # å·¦ä¾§æ¡†æ¶ï¼ˆä¸»å®¹å™¨ï¼‰
        left_frame = tk.Frame(self.root)
        left_frame.grid(row=0, column=0, sticky="nsew", padx=10, pady=10)
        left_frame.grid_rowconfigure(0, weight=0)  # è¾“å…¥åŒºåŸŸï¼ˆå›ºå®šé«˜åº¦ï¼‰
        left_frame.grid_rowconfigure(1, weight=1)  # å›å¤åŒºåŸŸï¼ˆå¯æ‰©å±•ï¼‰
        left_frame.grid_columnconfigure(0, weight=1)  # å•åˆ—å¡«å……

        # å³ä¾§æ¡†æ¶ - ä¸Šä¸‹ä¸¤éƒ¨åˆ†
        right_frame = tk.Frame(self.root)
        right_frame.grid(row=0, column=1, rowspan=2, sticky="nsew", padx=10, pady=10)
        right_frame.grid_rowconfigure(0, weight=1)  # å†å²è®°å½•åŒºåŸŸ
        right_frame.grid_rowconfigure(1, weight=0)  # ç©å®¶çŠ¶æ€åŒºåŸŸå›ºå®šé«˜åº¦
        right_frame.grid_columnconfigure(0, weight=1)

        # å³ä¸Šéƒ¨åˆ† - å†å²è®°å½•åŒºåŸŸ
        history_frame = tk.Frame(right_frame)
        history_frame.grid(row=0, column=0, sticky="nsew", pady=(0, 5))
        history_frame.grid_rowconfigure(1, weight=1)
        history_frame.grid_columnconfigure(0, weight=1)
        
        tk.Label(history_frame, text="å†å²è®°å½•ï¼š").grid(row=0, column=0, sticky="w")
        self.history_label = HTMLLabel(history_frame, html="", wrap=tk.WORD)
        self.history_label.grid(row=1, column=0, sticky="nsew")

        # å³ä¸‹éƒ¨åˆ† - ç©å®¶çŠ¶æ€è¡¨æ ¼
        status_frame = tk.Frame(right_frame, bd=2, relief=tk.GROOVE)
        status_frame.grid(row=1, column=0, sticky="nsew", pady=(5, 0))  
        status_frame.config(height=200)
        
        tk.Label(status_frame, text="ç©å®¶çŠ¶æ€", font=('Arial', 20, 'bold')).pack(pady=5)
        
        # åˆ›å»ºç©å®¶çŠ¶æ€è¡¨æ ¼
        self.player_status_table = tk.Frame(status_frame)
        self.player_status_table.pack(fill=tk.BOTH, expand=True, padx=5, pady=5)
        
        # åˆå§‹åŒ–çŠ¶æ€è¡¨æ ¼
        if self.test_mode:
            initial_status = {
                "ç”Ÿå‘½çŠ¶æ€": "æµ‹è¯•1æµ‹è¯•2æµ‹è¯•3æµ‹è¯•4æµ‹è¯•5æµ‹è¯•6æµ‹è¯•7æµ‹è¯•8æµ‹è¯•9æµ‹è¯•10æµ‹è¯•",
                "ææƒ§ç¨‹åº¦": "æµ‹è¯•1æµ‹è¯•2æµ‹è¯•3æµ‹è¯•4æµ‹è¯•5æµ‹è¯•6æµ‹è¯•7æµ‹è¯•8æµ‹è¯•9æµ‹è¯•10æµ‹è¯•",
                "ç”Ÿç†çŠ¶æ€": "æµ‹è¯•1æµ‹è¯•2æµ‹è¯•3æµ‹è¯•4æµ‹è¯•5æµ‹è¯•6æµ‹è¯•7æµ‹è¯•8æµ‹è¯•9æµ‹è¯•10æµ‹è¯•",
                "NPCé˜Ÿå‹": "æµ‹è¯•1æµ‹è¯•2æµ‹è¯•3æµ‹è¯•4æµ‹è¯•5æµ‹è¯•6æµ‹è¯•7æµ‹è¯•8æµ‹è¯•9æµ‹è¯•10æµ‹è¯•",
                "èƒŒåŒ…ç‰©å“": "æµ‹è¯•1æµ‹è¯•2æµ‹è¯•3æµ‹è¯•4æµ‹è¯•5æµ‹è¯•6æµ‹è¯•7æµ‹è¯•8æµ‹è¯•9æµ‹è¯•10æµ‹è¯•"
            }
        else:
            initial_status = {
                "ç”Ÿç†çŠ¶æ€": "å¥½çš„ä¸€æ¯”",
                "ææƒ§ç¨‹åº¦": "å®Œå…¨ä¸æ€‚",
                "NPCé˜Ÿå‹": "å­¤å®¶å¯¡äºº",
                "èƒŒåŒ…ç‰©å“": "èº«æ‰‹çº¸é’¥é’±",
                "å¯¹æ€ªç‰©çš„è®¤çŸ¥":"ä¸ˆäºŒå’Œå°š"
            }
        self.status = initial_status
        self.update_player_status(initial_status)

        # è¾“å…¥åŒºåŸŸ
        left_top_frame = tk.Frame(left_frame)
        left_top_frame.grid(row=0, column=0, sticky="nsew", padx=10, pady=(10, 5))
        left_top_frame.grid_columnconfigure(0, weight=1)
        
        # è¾“å…¥æ ‡ç­¾å’Œæ–‡æœ¬æ¡†
        tk.Label(left_top_frame, text="ç©å®¶è¾“å…¥ï¼š").grid(row=0, column=0, sticky="w")
        self.input_text = scrolledtext.ScrolledText(left_top_frame, wrap=tk.WORD, height=8)
        self.input_text.grid(row=1, column=0, sticky="ew")
        self.input_text.bind("<Return>", self.on_enter)

        # è¯­éŸ³æ§åˆ¶æŒ‰é’®åŒºåŸŸï¼ˆç§»åŠ¨åˆ°è¾“å…¥æ¡†ä¸‹æ–¹ï¼‰
        voice_control_frame = tk.Frame(left_top_frame)
        voice_control_frame.grid(row=2, column=0, sticky="e", pady=(5, 0))
        
        self.voice_toggle = tk.Button(
            voice_control_frame,
            text="ğŸ¤ å…³é—­", 
            command=self.toggle_voice_recognition,
            relief=tk.RAISED,
            width=8  # å›ºå®šæŒ‰é’®å®½åº¦
        )
        self.voice_toggle.pack(side=tk.RIGHT)
        
        # å›å¤æ˜¾ç¤ºåŒºåŸŸ
        left_bottom_frame = tk.Frame(left_frame)
        left_bottom_frame.grid(row=1, column=0, sticky="nsew", padx=10, pady=(5, 10))
        left_bottom_frame.grid_rowconfigure(1, weight=1)
        left_bottom_frame.grid_columnconfigure(0, weight=1)

        tk.Label(left_bottom_frame, text="ä¸»æŒäººï¼š").grid(row=0, column=0, sticky="w")
        self.reply_label = HTMLLabel(left_bottom_frame, html="", wrap=tk.WORD)
        self.reply_label.grid(row=1, column=0, sticky="nsew")

        # æŒ‰é’®åŒºåŸŸ
        button_frame = tk.Frame(self.root)
        button_frame.grid(row=1, column=0, sticky="ew", padx=10, pady=5)
        self.process_button = tk.Button(button_frame, text="ä¸‹ä¸€æ­¥", command=self.process_input)
        self.process_button.pack(side=tk.RIGHT)

        # å¤é€‰æ¡†
        self.checkbox_var = tk.BooleanVar(value=False)
        checkbox = tk.Checkbutton(
            self.root,
            text="æ–‡æœ¬æœ—è¯»",
            variable=self.checkbox_var,
            command=self.on_checkbox_change
        )
        checkbox.grid(row=2, column=0, columnspan=2, sticky="w", padx=10, pady=5)
        
        # çŠ¶æ€æ 
        self.status_var = tk.StringVar()
        self.status_var.set("å‡†å¤‡å°±ç»ª")
        status_label = tk.Label(self.root, textvariable=self.status_var, bd=1, relief=tk.SUNKEN, anchor=tk.W)
        status_label.grid(row=3, column=0, columnspan=2, sticky="ew")

    def window_init(self):
        # åˆå§‹åŒ–çª—å£å†…å®¹
        if self.test_mode == False:
            reply = self.agentManager.show_background()
            reply_txt = cls_GeneralTools.markdown_to_text_simple(reply)
        else:
            reply = ''
            reply_txt = ''

        initial_input = "# æ¬¢è¿ä½¿ç”¨\nç©å®¶åœ¨è¿™é‡Œè¾“å…¥..."
        self.input_text.insert(tk.END, initial_input)

        initial_bottom_content = "# è¿™é‡Œæ˜¯æ¯è½®ä¸»æŒäººçš„å›å¤ï¼Œæœ¬åœºæ¸¸æˆçš„æ¸¸æˆè§„åˆ™å¦‚ä¸‹ï¼š\n" + reply  
        initial_bottom_html = markdown.markdown(initial_bottom_content)
        self.reply_label.set_html(initial_bottom_html)

        initial_right_content = "**è¿™é‡Œä¼šæ˜¾ç¤ºæ‰€æœ‰çš„å†å²å¯¹è¯è®°å½•**"
        initial_right_html = markdown.markdown(initial_right_content)
        self.history_label.set_html(initial_right_html)
        
        self.root.state('zoomed')
        if self.checkbox_var.get():
            print('reading')
            self.voiceManager.speak(reply_txt)

    def on_enter(self, event):
        if event.state & 0x1:  # æ£€æŸ¥ Shift é”®æ˜¯å¦è¢«æŒ‰ä¸‹
            return
        self.process_input()
        return "break"

    def process_input(self, history=''):
        if self.streaming:
            return
            
        input_content = self.input_text.get("1.0", tk.END).strip()
        if not input_content:
            return
            
        self.streaming = True
        self.full_response = ""
        self.process_button.config(state=tk.DISABLED)
        self.status_var.set("æ­£åœ¨è·å–å›å¤...")
        
        Thread(target=self.fetch_stream_response, args=(input_content, history), daemon=True).start()

    def fetch_stream_response(self, input_content, history):
        stream_flag = True
        try:
            response = self.agentManager.talk_2_kp(prompt=input_content, stream_mode=stream_flag)
            self.root.after(0, lambda: self.safe_update_status("æ­£åœ¨æ¥æ”¶å›å¤..."))  
            
            for chunk in response:
                if not self.streaming:
                    break
                    
                if hasattr(chunk, 'choices') and chunk.choices[0].delta.content:
                    content = chunk.choices[0].delta.content
                    self.full_response += content
                    self.root.after(0, lambda c=content: self.safe_update_reply(c)) 
            
            if self.streaming:
                self.root.after(0, lambda: self.finalize_response(history))  
                
        except Exception as e:
            self.root.after(0, lambda: self.safe_set_error(f"å‘ç”Ÿé”™è¯¯: {str(e)}"))  
        finally:
            self.streaming = False
            self.root.after(0, lambda: self.safe_update_buttons())  

    def finalize_response(self, history):
        reply_txt = cls_GeneralTools.markdown_to_text_simple(self.full_response,test_flag=False)
        self.agentManager.kp_history.append(self.agentManager.pack_content('assistant', reply_txt))
        # print(f'æ‰“åŒ…å›ä¼ :{len(self.agentManager.kp_history)}')

        if self.checkbox_var.get():
            print('reading')
            self.voiceManager.speak(reply_txt)
        
        html_content = markdown.markdown(history + "\n\n" + self.full_response)

        self.safe_update_status("æ­£åœ¨åŠ è½½å†å²è®°å½•...")
        self.safe_update_history(html_content)
        
        self.safe_update_status("æ­£åœ¨æ›´æ–°ç©å®¶çŠ¶æ€...")
        new_status_str = self.agentManager.json_reply(self.status)
        print(repr(new_status_str))
        new_status_js = cls_GeneralTools.parse_status_json(new_status_str)
        if new_status_js is None:
            new_status_js =self.status
        self.update_player_status(new_status_js)
        self.status = new_status_js

        self.safe_update_status("å›å¤æ¥æ”¶å®Œæˆ")

    def update_player_status(self, status_data):
        """é€šè¿‡JSONæ•°æ®æ›´æ–°æ•´ä¸ªçŠ¶æ€è¡¨æ ¼"""
        # æ¸…ç©ºç°æœ‰çŠ¶æ€é¡¹
        for widget in self.player_status_table.winfo_children():
            widget.destroy()
        # æ·»åŠ æ–°çš„çŠ¶æ€é¡¹
        for name, value in status_data.items():
            self._add_single_status_item(name, value)

    def _add_single_status_item(self, name, value):
        """å†…éƒ¨æ–¹æ³•ï¼šæ·»åŠ å•ä¸ªçŠ¶æ€é¡¹åˆ°è¡¨æ ¼ï¼Œè‡ªé€‚åº”å®½åº¦"""
        row = tk.Frame(self.player_status_table)
        row.pack(fill=tk.X, pady=2)
        
        # æ ‡ç­¾ä¸å†å›ºå®šå®½åº¦ï¼Œæ”¹ä¸ºæ°´å¹³å¡«å……
        tk.Label(row, text=f"{name}:", anchor='w', font=('Arial', 9)).pack(side=tk.LEFT, fill=tk.X)
        tk.Label(row, text=value, anchor='w', font=('Arial', 9)).pack(side=tk.LEFT, fill=tk.X)

    def safe_update_reply(self, content):
        self.root.after(0, lambda: self.update_reply(content))

    def safe_update_history(self, html_content=None):
        """æ›´æ–°å†å²è®°å½•æ˜¾ç¤ºï¼Œä»ç¬¬äºŒæ¡å¼€å§‹æ˜¾ç¤ºkp_historyå†…å®¹"""
        history_text = ""
        
        for i, message in enumerate(self.agentManager.kp_history):
            role = message.get("role", "unknown")
            match role:
                case 'system': role = 'ç³»ç»Ÿ'
                case 'user': role = 'ç©å®¶'
                case 'assistant': role = 'ä¸»æŒäºº'
                case _: pass
            
            if i < 3:  # è·³è¿‡å‰ä¸¤æ¡system prompt
                continue
                
            content = cls_GeneralTools.markdown_to_text_simple(message.get("content", ""))
            separator = "\n" + "â”€" * 50 + "\n"
            history_text += f"{separator}{role}:\n{content}\n"
        
        html_content = markdown.markdown(history_text.replace("\n", "<br>"))
        styled_html = f"""
        <div style='
            font-family: Arial, sans-serif;
            font-size: 14px;
            line-height: 1.6;
            padding: 10px;
        '>
            {html_content}
        </div>
        """
        self.root.after(0, lambda: self.history_label.set_html(styled_html))

    def safe_update_status(self, message):
        self.root.after(0, lambda: self.status_var.set(message))

    def safe_set_error(self, error_msg):
        self.root.after(0, lambda: self.reply_label.set_html(
            f"<div style='color: red; padding: 10px;'>{error_msg}</div>"
        ))

    def safe_update_buttons(self):
        self.root.after(0, lambda: self.process_button.config(state=tk.NORMAL))

    def update_reply(self, content):
        try:
            html_content = markdown.markdown(self.full_response)
            styled_html = f"""
            <div style='
                font-family: Arial, sans-serif;
                font-size: 14px;
                line-height: 1.6;
                padding: 10px;
            '>
                {html_content}
            </div>
            """
            self.reply_label.set_html(styled_html)
            self.reply_label.see(tk.END)
            self.status_var.set(f"æ¥æ”¶ä¸­... {len(self.full_response)} å­—ç¬¦")
        except Exception as e:
            self.reply_label.set_html(
                f"<div style='color: red; padding: 10px;'>æ¸²æŸ“é”™è¯¯: {str(e)}</div>"
            )

    def on_checkbox_change(self):
        print("å¤é€‰æ¡†çŠ¶æ€å·²å˜åŒ–:", self.checkbox_var.get())
        
    def toggle_voice_recognition(self):
        """åˆ‡æ¢è¯­éŸ³è¯†åˆ«çŠ¶æ€"""
        if not self.vosk_model:
            self.status_var.set("è­¦å‘Šï¼šVoskè¯­éŸ³è¯†åˆ«ä¸å¯ç”¨ï¼ˆæœªæ­£ç¡®åˆå§‹åŒ–ï¼‰")
            return
            
        was_active = self.voice_recognition_active
        self.voice_recognition_active = not was_active
        
        if self.voice_recognition_active:
            self.audio_stream.start_stream()
            self.voice_toggle.config(text="ğŸ¤ å¼€å¯", relief=tk.SUNKEN, bg="#ffcccc")
            self.status_var.set("è¯­éŸ³è¯†åˆ«å·²å¼€å¯ï¼Œè¯·å¼€å§‹è¯´è¯...")
        else:
            self.audio_stream.stop_stream()
            self.voice_toggle.config(text="ğŸ¤ å…³é—­", relief=tk.RAISED, bg="SystemButtonFace")
            self.status_var.set("è¯­éŸ³è¯†åˆ«å·²å…³é—­")
            
            # å…³é—­æ—¶æ¸…é™¤æ‰€æœ‰ç©ºæ ¼ï¼ˆåŒ…æ‹¬å•è¯é—´çš„æ­£å¸¸ç©ºæ ¼ï¼‰
            current_text = self.input_text.get("1.0", tk.END)
            cleaned_text = current_text.replace(" ", "")  # ç§»é™¤æ‰€æœ‰ç©ºæ ¼
            self.input_text.delete("1.0", tk.END)
            self.input_text.insert("1.0", cleaned_text.strip())  # ä¿ç•™æ¢è¡Œç¬¦ç­‰å…¶ä»–ç©ºç™½å­—ç¬¦

    def append_to_input(self, text):
        """å°†è¯†åˆ«åˆ°çš„æ–‡æœ¬è¿½åŠ åˆ°è¾“å…¥æ¡†"""
        current = self.input_text.get("1.0", tk.END).strip()
        if current.endswith(("ã€‚", "ï¼", "ï¼Ÿ", ".", "!", "?")) or not current:
            text = text.capitalize()  # æ–°å¥å­é¦–å­—æ¯å¤§å†™
            
        self.input_text.insert(tk.END, text)
        self.input_text.see(tk.END)  # æ»šåŠ¨åˆ°æœ«å°¾

    def __del__(self):
        """æ¸…ç†èµ„æº"""
        if hasattr(self, 'audio_stream'):
            self.audio_stream.stop_stream()
            self.audio_stream.close()
        if hasattr(self, 'pyaudio'):
            self.pyaudio.terminate()
    

    def on_window_close(self):
        """çª—å£å…³é—­æ—¶çš„å¤„ç†"""
        try:
            # ä¿å­˜å†å²è®°å½•
            self.save_history_to_file()
            
            # æ¸…ç†èµ„æº
            if hasattr(self, 'audio_stream'):
                self.audio_stream.stop_stream()
                self.audio_stream.close()
            if hasattr(self, 'pyaudio'):
                self.pyaudio.terminate()
                
            # å…³é—­çª—å£
            self.root.destroy()
        except Exception as e:
            print(f"å…³é—­çª—å£æ—¶å‡ºé”™: {e}")
            self.root.destroy()

    def save_history_to_file(self):
        """å°†å¯¹è¯å†å²ä¿å­˜åˆ°æ–‡ä»¶"""
        try:
            if not hasattr(self.agentManager, 'kp_history'):
                return
                
            filename = f"TRPG_å¯¹è¯è®°å½•_{time.strftime('%Y%m%d_%H%M%S')}.txt"
            
            with open(filename, 'w', encoding='utf-8') as f:
                for i, message in enumerate(self.agentManager.kp_history):
                    role = message.get("role", "unknown")
                    content = message.get("content", "")
                    
                    # æ ¼å¼åŒ–è§’è‰²åç§°
                    if role == 'system': role = 'ç³»ç»Ÿ'
                    elif role == 'user': role = 'ç©å®¶'
                    elif role == 'assistant': role = 'ä¸»æŒäºº'
                    
                    # å†™å…¥æ–‡ä»¶
                    f.write(f"ã€{role}ã€‘\n")
                    f.write(f"{content}\n")
                    f.write("-"*50 + "\n\n")
                    
            print(f"å¯¹è¯å†å²å·²ä¿å­˜åˆ°: {filename}")
        except Exception as e:
            print(f"ä¿å­˜å†å²è®°å½•å¤±è´¥: {e}")
```
